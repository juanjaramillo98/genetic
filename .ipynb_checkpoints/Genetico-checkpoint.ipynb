```python
def getModel():
    model = Sequential()
    model.add(Dense(4, input_shape=(4,)))
    model.add(Activation('relu'))
    model.add(Dense(9, input_shape=(4,)))
    model.add(Activation('relu'))
    model.add(Dense(3, input_shape=(9,)))
    model.add(Activation('softmax'))
    
    model.compile(loss = 'mse', optimizer = 'adam', metrics=['accuracy'] )
    return model
    
def getModel2():
    model = Sequential()
    model.add(Dense(5, input_shape=(5,)))
    model.add(Activation('relu'))
    model.add(Dense(9, input_shape=(4,)))
    model.add(Activation('relu'))
    model.add(Dense(1, input_shape=(9,)))
    model.add(Activation('sigmoid'))
    
    model.compile(loss = 'mse', optimizer = 'adam', metrics=['accuracy'] )
    return model
    
def save_pool(a):
    for xi in range(total_models):
        current_pool[xi][0].save_weights("SavedModels/decicion" + str(xi) +"_"+ a +".keras")
        current_pool[xi][1].save_weights("SavedModels/cantidad" + str(xi) +"_"+ a +".keras")
    print("Saved current pool!")

def predict_action(EntradasDelModelo, model_num):
    global current_pool
    # se normalizan las entradas del modelo
    
    neural_input = EntradasDelModelo
    output_prob = current_pool[model_num][0].predict(neural_input,batch_size=400)
    accion = np.argmax(output_prob,axis =1)
    
    return accion

def predict_cantidad(EntradasDelModelo, model_num):
    global current_pool
    # se normalizan las entradas del modelo
    neural_input = EntradasDelModelo
    output_prob = current_pool[model_num][1].predict(neural_input,batch_size=400)
    
    return output_prob

def model_crossover(parent1, parent2):
    # obtain parent weights
    # get random gene
    # swap genes
    global current_pool

    weight1 = current_pool[parent1][0].get_weights()
    weight2 = current_pool[parent2][0].get_weights()

    new_weight1 = weight1
    new_weight2 = weight2

    gene = random.randint(0,len(new_weight1)-1)

    new_weight1[gene] = weight2[gene]
    new_weight2[gene] = weight1[gene]

    return np.asarray([new_weight1,new_weight2])

def model_crossover_brain(parent1, parent2):
    # obtain parent weights
    # get random gene
    # swap genes
    global current_pool

    weight1 = current_pool[parent1][1].get_weights()
    weight2 = current_pool[parent2][1].get_weights()

    new_weight1 = weight1
    new_weight2 = weight2

    gene = random.randint(0,len(new_weight1)-1)

    new_weight1[gene] = weight2[gene]
    new_weight2[gene] = weight1[gene]

    return np.asarray([new_weight1,new_weight2])



def model_mutate(weights):#,generation):
    
    # mutate each models weights
    for i in range(len(weights)):
        for j in range(len(weights[i])):
            if( random.uniform(0,1) > 1-mutation_rate):
                change = random.uniform(-0.3,0.3)
                weights[i][j] += change
    return weights

def ejecutar_accion(accion,price,model_num,cantidad):
    
    #0  es mantenerse
    #1  es comprar
    #2  es vender

    if(accion == 0):
        return
    
    elif(accion == 1):
        capacidad_compra = int(wallet[model_num]/price)
        cantidad_a_comprar = int(capacidad_compra*cantidad)
        wallet[model_num] = wallet[model_num]-price*cantidad_a_comprar
        portafolio[model_num] = portafolio[model_num] + cantidad_a_comprar 
        
    else:
        if portafolio[model_num] == 0:
            return
        capacidad_venta = portafolio[model_num]
        cantidad_a_vender = int(capacidad_venta*cantidad)
        portafolio[model_num] = portafolio[model_num] - cantidad_a_vender
        wallet[model_num] = wallet[model_num] + price*cantidad_a_vender          
    return
```

```python
load_saved_pool = False
save_current_pool = True
total_models = 100
mutation_rate = 0.10
wallet_inicial = 100000
longitud_simulacion = 100
acciones_iniciales = 0
```

```python
current_pool = []
fitness = []
wallet = []
portafolio = []
historicoW = []
historicoP = []

portafolio_inicial = int(int(wallet_inicial/precio[momento])*acciones_iniciales)
wallet_inicial = wallet_inicial - (precio[momento]*portafolio_inicial)

for i in range(total_models):
    current_pool.append([getModel(),getModel2()])
    wallet.append(wallet_inicial)
    fitness.append(0)
    portafolio.append(portafolio_inicial)
    historicoW.append([])
    historicoP.append([])
```

```python

for k in range(5):
    
    #--------------------------------- simulamos 400 dias en el mercado en un punto aleatorio del dataset ----------
    ejecucion = time()
    for i in range(total_models):
        actions = predict_action(imput[momento:momento+longitud_simulacion],i)
        
        imput_plus = np.append(imput[momento:momento+longitud_simulacion],(actions-1).reshape(longitud_simulacion,1),axis=1)
        cantidades = predict_cantidad(imput_plus,i)
        
        for j in range(momento , momento+longitud_simulacion):
            ejecutar_accion(actions[j-momento],precio[j],i,cantidades[j-momento])
            
#             if wallet[i]>wallet_inicial:
#                 fitness[i]+=1
            historicoW[i].append(wallet[i])  
            historicoP[i].append(portafolio[i])


    #-------------------------------- evaluamos a los modelos -------------------------------------------        

    for i in range(total_models):

        total_activos = wallet[i]+(portafolio[i]*(precio[momento+longitud_simulacion]*0.7))

        if total_activos != wallet_inicial:
            fitness[i] = total_activos 
        else:
            fitness[i] =  0

    mejores = np.argsort(fitness)
#     print(mejores[-1])
#     print(wallet[mejores[-1]]+(portafolio[mejores[-1]]*(precio[momento+400])))
#     print(mejores[-2])
#     print(wallet[mejores[-2]]+(portafolio[mejores[-2]]*(precio[momento+400])))

    print('generacion:'+str(k))
    print('dinero: '+str(wallet[mejores[-1]]+(portafolio[mejores[-1]]*(precio[momento+longitud_simulacion]))))
    #------------------------------- los 2 mejores se reproducen y crean la nueva generacion ---------------


    for k in range(0,total_models,2):
         
        pesos = model_crossover(mejores[-1], mejores[-2])
        pesos_brain = model_crossover_brain(mejores[-1], mejores[-2])

        pesos1 = pesos[0]
        pesos2 = pesos[1]
        
        pesos1 = model_mutate(pesos1)
        pesos2 = model_mutate(pesos2)
        
        pesos1B = pesos_brain[0]
        pesos2B = pesos_brain[1]
        
        pesos1B = model_mutate(pesos1B)
        pesos2B = model_mutate(pesos2B)
        
        if k == 0:
            current_pool[k][0].set_weights(current_pool[mejores[-1]][0].get_weights())
            current_pool[k][1].set_weights(current_pool[mejores[-1]][1].get_weights())
            
            current_pool[k+1][0].set_weights(current_pool[mejores[-2]][0].get_weights())
            current_pool[k+1][1].set_weights(current_pool[mejores[-2]][1].get_weights())
            
        if k == 2:
            current_pool[k][0].set_weights(current_pool[mejores[-2]][0].get_weights())
            current_pool[k][1].set_weights(current_pool[mejores[-1]][1].get_weights())
            
            current_pool[k+1][0].set_weights(current_pool[mejores[-1]][0].get_weights())
            current_pool[k+1][1].set_weights(current_pool[mejores[-2]][1].get_weights())
            
        current_pool[k][0].set_weights(pesos1)
        current_pool[k+1][0].set_weights(pesos2)
        
        current_pool[k][1].set_weights(pesos1B)
        current_pool[k+1][1].set_weights(pesos2B)
        

    #--------------------------------------imprimo los historicos----------------------------------------------
    historicoWCopy = np.array(historicoW)
    historicoPCopy = np.array(historicoP)
    
    preciopd[momento:momento+longitud_simulacion].plot(figsize=(15,3))
    plt.show()
    
    plt.figure(figsize=(15,3))
    plt.plot(historicoWCopy[mejores[-1]])
    plt.plot(historicoWCopy[mejores[-2]])
    plt.plot(historicoWCopy[mejores[-3]])
    plt.plot(historicoWCopy[mejores[-4]])
    plt.show()
    
    plt.figure(figsize=(15,3))
    plt.plot(historicoWCopy.T)
    plt.show()
    
    plt.figure(figsize=(15,3))
    plt.plot(historicoPCopy.T)
    plt.show()

    #------------------------------------- reinicio wallets y demas --------------------------------------------
    a = historicoW
    b = historicoP
    historicoW = []
    historicoP = []
    for i in range(total_models):    
        wallet[i]=wallet_inicial 
        fitness[i] = 0
        portafolio[i] = portafolio_inicial
        historicoW.append([])
        historicoP.append([])


    ejecucion = time()-ejecucion
    print('tiempo: '+str(ejecucion))
```
